##  Situation of covid 19 in Mexico


#### Data loading and wrangling

```python
#Importar librerias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import time
import datetime
import json
import folium
```

```python
#Cargar datos
dfm = pd.read_csv("CoordMunicipios.csv", encoding = "ISO-8859-1")
dfe = pd.read_csv("CoordEstados.csv", encoding = "ISO-8859-1")
df = pd.read_csv("200504COVID19MEXICO.csv", encoding = "ISO-8859-1")
```

```python
#Reemplace por  Nans y 0,1:
cols=["SEXO","RESULTADO","EPOC","EMBARAZO","INTUBADO","DIABETES","ASMA","INMUSUPR","HIPERTENSION",'OTRA_COM','CARDIOVASCULAR', 'OBESIDAD', 'RENAL_CRONICA', 'TABAQUISMO',
       'OTRO_CASO','UCI']
df[cols]=df[cols].replace([3,97,98,99],np.nan)
df[cols]=df[cols].replace(2,0)
```

```python
#Creo columna MUERTES: Muertes 0,1(1=Muerte)
lista_def=[]
for i in df["FECHA_DEF"]:
    if i=="9999-99-99":
        lista_def.append(0)
    else:
        lista_def.append(1)
df["MUERTES"]=lista_def
```

```python
#Creo columna ENF: Con enferm!=0, Sin enferm=0
#NEUMONIA no entra
cols_enf=df[['EPOC','DIABETES','ASMA','INMUSUPR','HIPERTENSION',"EMBARAZO",
   'OTRA_COM','CARDIOVASCULAR','OBESIDAD','RENAL_CRONICA','TABAQUISMO',]]
df['ENF'] = cols_enf.sum(axis=1)
```

```python
#Cambio a date time cols fechas
df["FECHA_DEF"]= df["FECHA_DEF"].replace("9999-99-99",np.nan)
from datetime import datetime
cols_fechas=['FECHA_INGRESO', 'FECHA_SINTOMAS',"FECHA_DEF"]
df[cols_fechas]=df[cols_fechas].apply(pd.to_datetime, format="%d/%m/%Y")
```

```python
#Creo columns tiempo en UCI
df["TIEMPO_UCI"]=df["FECHA_DEF"]- df["FECHA_INGRESO"]
#Creo columna tiempo hasta ingreso
df["TIEMPO_INGRESO"]=df["FECHA_INGRESO"]-df["FECHA_SINTOMAS"]
df["TIEMPO_TOT"]=df["FECHA_DEF"]-df["FECHA_SINTOMAS"]
```

```python
#Unir dataframes
df= df.rename(columns={"MUNICIPIO_RES":"Clave Municipio","ENTIDAD_RES":"Clave Estado"})
df= pd.merge(left=df, right=dfm, on =["Clave Municipio","Clave Estado"])
```

```python
#Numero total de pacientes:
print("TOTAL PATIENTS: ",df.shape[0])
print("POSITIVES: ",df[df["RESULTADO"]==1].shape[0])
print("NEGATIVES: ",df[df["RESULTADO"]==0].shape[0])
print("POSITIVES IN IUC: ",df[(df["UCI"]==1)&(df["RESULTADO"]==1)].shape[0])
print("POSITIVES VENTILATOR REQUIRED: ",df[(df["RESULTADO"]==1)&(df["INTUBADO"]==1)].shape[0])
print("FIRST DATE OF ENTRY: ", df["FECHA_INGRESO"].min())
print("LAST DATE OF ENTRY: ", df["FECHA_INGRESO"].max())
print("POSITIVE TOTAL_DEATHS: ",df[(df["RESULTADO"]==1)&(df["MUERTES"]==1)].shape[0])
print("AVERAGE TIME IUC (POSITIVES DEATHS): ",df[df["RESULTADO"]==1]['TIEMPO_UCI'].mean())
print("MAX TIME IUC (POSITIVES DEATHS): ",df[df["RESULTADO"]==1]['TIEMPO_UCI'].max())
print("RISK FACTORS (POSITIVES): ",df[(df["RESULTADO"]==1)&(df["ENF"]!=0)].shape[0])
print("NO RISK FACTORS (POSITIVES): ",df[(df["RESULTADO"]==1)&(df["ENF"]==0)].shape[0])
print("RISK FACTORS (POSITIVES DEATHS): ",df[(df["RESULTADO"]==1)&(df["ENF"]!=0)&(df["MUERTES"]==1)].shape[0])
print("NO RISK FACTORS (POSITIVES DEATHS): ",df[(df["RESULTADO"]==1)&(df["ENF"]==0)&(df["MUERTES"]==1)].shape[0])
```

```python
#FILTRADO CON O SIN FR:
#GRAFICA BAR
pos_con=df[(df["RESULTADO"]==1)&(df["ENF"]!=0)]["RESULTADO"].shape[0]
pos_sin=df[(df["RESULTADO"]==1)&(df["ENF"]==0)]["RESULTADO"].shape[0]
lista_pos=[pos_con,pos_sin]

muerte_con=df[(df["RESULTADO"]==1)&(df["MUERTES"]==1)&(df["ENF"]!=0)]["RESULTADO"].shape[0]
muerte_sin=df[(df["RESULTADO"]==1)&(df["MUERTES"]==1)&(df["ENF"]==0)]["RESULTADO"].shape[0]
lista_muerte=[muerte_con,muerte_sin]

intu_con=df[(df["RESULTADO"]==1)&(df["INTUBADO"]==1)&(df["ENF"]!=0)]["RESULTADO"].shape[0]
intu_sin=df[(df["RESULTADO"]==1)&(df["INTUBADO"]==1)&(df["ENF"]==0)]["RESULTADO"].shape[0]
lista_intu=[intu_con,intu_sin]

uci_con=df[(df["RESULTADO"]==1)&(df["UCI"]==1)&(df["ENF"]!=0)]["RESULTADO"].shape[0]
uci_sin=df[(df["RESULTADO"]==1)&(df["UCI"]==1)&(df["ENF"]==0)]["RESULTADO"].shape[0]
lista_uci=[uci_con,uci_sin]

df5=pd.DataFrame(columns=["POSITIVES","VENTILATOR","DEATHS","IUC"], index=["RISK FACTOR", "NO RISK FACTORS"])
df5["POSITIVES"]=lista_pos
df5["VENTILATOR"]=lista_intu
df5["DEATHS"]=lista_muerte
df5["IUC"]=lista_uci
```

```python
#GRAFICA BAR
fig, ax = plt.subplots(figsize=(8,8))
etiquetas=["RISK FACTORS", "NO RISK FACTORS"]
valores=df5["POSITIVES"]
valores1=df5["IUC"]
valores2=df5["DEATHS"]
valores3=df5["VENTILATOR"]

x = np.arange(len(etiquetas))  
width=0.4

plt.bar(x - width/8,valores2,0.1,color="#ff0000", label = "DEATHS")
plt.bar(x - width/2.5,valores,0.1, color="#890000", label = "POSITIVES")
plt.bar(x + width/2.4,valores1,0.1, color="#ffb1b1", label = "IUC")
plt.bar(x + width/7,valores3,0.1, color="#140000", label = "VENTILATOR")

plt.legend(loc="best") 
ax.set_xticks(x)
ax.set_xticklabels(etiquetas)
plt.show()

#plt.bar(x - width/2,valores,0.4, color="blue")
#plt.bar(x + width/2,valores1,0.4 ,color="red")
```

```python
#GRAFICA PIE
#Lista con num de intubados por factor de riesgo:
df_int=df[(df["RESULTADO"]==1)&(df["INTUBADO"]==1)]
lista_count=[]
lista_enf=['EPOC','DIABETES','ASMA','INMUSUPR','HIPERTENSION',
   'OTRA_COM','CARDIOVASCULAR','OBESIDAD','RENAL_CRONICA','TABAQUISMO']
for i in lista_enf:
    lista_count.append(df_int[df_int[i]==1][i].sum())
```

```python
#GRAFICA PIE 
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(aspect="equal"))

labels = lista_enf
sizes = lista_count

explode = (0, 0.1, 0, 0, 0.1, 0, 0, 0.1, 0, 0,)
colores=["#FFFACD", "#FF7F50", "#FFFACD", "#FFFACD", "#FF7F50", "#FFFACD", "#FFFACD", "#FF7F50", "#FFFACD", "#FFD700"]

ax.pie(sizes, explode=explode, labels=labels,colors=colores, autopct='%1.1f%%',shadow=True, startangle=90)

ax.axis('equal')
plt.title("VENTILATOR REQUIRED")
plt.tight_layout()
```

```python
df_pos=df[df["RESULTADO"]==1]
#HISTOGRAMA
plt.subplots(figsize=(8,6))
sns.distplot(df_pos[(df_pos["ENF"]!=0)&(df_pos["INTUBADO"]==1)]["EDAD"], kde_kws={"color": "r", "lw": 2, "label": "Previous risk factors"},hist=False)
sns.distplot(df_pos[(df_pos["ENF"]==0)&(df_pos["INTUBADO"]==1)]["EDAD"], kde_kws={"color": "b", "lw": 2, "label": "No previous risk factors "},hist=False)
plt.title("VENTILATOR REQUIRED / AGE")
plt.xlabel("AGE")
plt.show()
```

```python
#filtro positivos e intubados
dfpi=df[(df["RESULTADO"]==1)&(df["INTUBADO"]==1)]

#filtro edad
HYPERTENSION= dfpi[dfpi["HIPERTENSION"]==1]["EDAD"]
DIABETES= dfpi[dfpi["DIABETES"]==1]["EDAD"]
OBESITY= dfpi[dfpi["OBESIDAD"]==1]["EDAD"]

#boxplots
all_data=[HYPERTENSION,DIABETES,OBESITY]
labels=["HYPERTENSION","DIABETES","OBESITY"]

fig, ax = plt.subplots(figsize=(15,5))

ax.boxplot(all_data, labels=labels)
plt.ylabel("Age")
plt.title("VENTILATOR / AGE")
plt.show()
            
```

#### Conclusions

People who have any of the mentioned risk factors are more predisposed to both infection and the complication of the disease and may require a ventilator and ICU.

The mortality rate is significantly higher in people with these risk factors.

The risk of having respiratory complications increases considerably in patients older than 40 years but no significant differences are observed between those who have previous risk factors or not. 

Although it is more common to develop this type of pathology in older people and the average age in intubated patients is somewhat older.

Analyzing more in depth the 3 critical diseases for the complication of COVID, it is observed that in the case of obesity the average risk age decreases a little with respect to hypertension and diabetes.

### Evolution of Covid

```python
#Filtrado y wrangling de MUERTES:
df2=df[(df["RESULTADO"]==1)&df["MUERTES"]==1]
df2=df2.sort_values(by='FECHA_DEF')
df2=df2.groupby("FECHA_DEF").count().iloc[:,1:2]
df2=df2.asfreq(freq="d")
#Creo col con suma ac
muertes=df2["ID_REGISTRO"]
total_m=muertes.cumsum()
df2["CUMSUM"]=total_m
#Filtado y wrang POSITIVOS:
df3=df[df["RESULTADO"]==1]
df3=df3.sort_values(by='FECHA_INGRESO')
df3=df3[df3["FECHA_INGRESO"]> "2020-03-17"]
df3=df3.groupby("FECHA_INGRESO").count().iloc[:,1:2]
df3=df3.asfreq(freq="d")
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(20,12))
plt.subplots_adjust(wspace=None, hspace=0.5)
ax1=df2.ID_REGISTRO.plot(ax=axes[0],title="Positives evolution",color="red")
ax1.set_xlabel("Date")
ax1.set_ylabel("Num positives")
ax2=df3.ID_REGISTRO.plot(ax=axes[1],title="Deaths evolution")
ax2.set_xlabel("Date")
ax2.set_ylabel("Num deaths")
plt.show()
```

```python
df2.CUMSUM.plot(figsize=(20,5),kind="bar",color="darkblue")
plt.xlabel("Date")
plt.ylabel("Number of deaths")
plt.title("Evolution of death rate")
plt.show()
```

#### Conclusions

The positives analyzed are only those are at hospitals.
It is observed that the number of positives and deaths increased a lot in the first week of April, 
and it peaked at the end of that month.
The total deaths begin to stabilize in the end of the month.

### Maps and tables

```python
#Preprocesado mapas
df_pos=df[df["RESULTADO"]==1]
pos=df_pos.groupby(["Clave Estado"]).sum()["RESULTADO"].tolist()
muer=df_pos.groupby(["Clave Estado"]).sum()["MUERTES"].tolist()

df_map=dfe[0:32]

df_map["MUERTES"]=muer
df_map["POSITIVOS"]=pos
df_map["POSITIVOS"]=df_map["POSITIVOS"].astype(int)

#Reajuste de nombres json
df_map=df_map.replace("Ciudad De México",'CDMX')
df_map=df_map.replace("Querétaro",'Queretaro')
df_map=df_map.replace("Estado De México","Estado De Mexico")
df_map=df_map.replace("Yucatán","Yucatan")
df_map=df_map.replace("Michoacán De Ocampo","Michoacan De Ocampo")
df_map=df_map.replace("San Luis Potosí","San Luis Potosi")
df_map=df_map.replace("Nuevo León","Nuevo Leon")
```

```python
latitude=23.634501
longitude=-102.552784
map_pos = folium.Map(location=[latitude, longitude], zoom_start=5) # generate map centred around the Conrad Hotel

for lat, lng, label in zip(df_map.Latitud, df_map.Longitud, df_map.POSITIVOS):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(map_pos)

# display map
print("CLICK AND CHECK THE POSITIVES")
map_pos
```

```python
latitude=23.634501
longitude=-102.552784
map_pos = folium.Map(location=[latitude, longitude], zoom_start=5) # generate map centred around the Conrad Hotel

for lat, lng, label in zip(df_map.Latitud, df_map.Longitud, df_map.MUERTES):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='red',
        popup=label,
        fill = True,
        fill_color='red',
        fill_opacity=0.6
    ).add_to(map_pos)

# display map
print("CLICK AND CHECK THE DEATHS")
map_pos
```

```python
#Municipio con max de positivos por estado:
df_pos=df[df["RESULTADO"]==1]
df_pos=df_pos.groupby(["Estado","Municipio"],as_index=False).sum()
lista_unicos=df["Estado"].unique()
lista_mun=[]
lista_values=[]
for i in lista_unicos:
    lista_mun.append(df_pos[df_pos["Estado"]==i].max()[1])
    lista_values.append(df_pos[df_pos["Estado"]==i].max()[2])
df_max=pd.DataFrame(columns=["State","Region","Positives"])
df_max["State"]=lista_unicos
df_max["Region"]=lista_mun
df_max["Positives"]=lista_values
df_max.set_index('State', inplace=True)
df_max=df_max.sort_values(by="Positives",ascending=False)

print("CRITICAL AREAS")
df_max
```

```python
#Municipio con max de positivos por estado:
df_muer=df[(df["RESULTADO"]==1)&(df["MUERTES"]==1)]
df_muer=df_muer.groupby(["Estado","Municipio"],as_index=False).sum()
lista_unicos=df["Estado"].unique()
lista_mun=[]
lista_values=[]
for i in lista_unicos:
    lista_mun.append(df_muer[df_muer["Estado"]==i].max()[1])
    lista_values.append(df_muer[df_muer["Estado"]==i].max()[2])
df_max2=pd.DataFrame(columns=["State","Region","Deaths"])
df_max2["State"]=lista_unicos
df_max2["Region"]=lista_mun
df_max2["Deaths"]=lista_values
df_max2.set_index('State', inplace=True)
df_max2=df_max2.sort_values(by="Deaths",ascending=False)
print("CRITICAL AREAS")
df_max2
```
